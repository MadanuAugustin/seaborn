```python
# creating a scatter plot using seaborn
# importing the libraries
import seaborn as sns

# loading the dataset
mydataset = sns.load_dataset('fmri')
mydataset

sns.scatterplot(x='timepoint', y='signal', data = mydataset, hue = 'region', style='event')
```

```python
# visualizing the relationship between two variables using relplot method
# importing the libraries
import seaborn as sns
import pandas as pd

# loading the dataset
mydata = pd.read_csv('FuelConsumption.csv')

# visualizing the data using relplot
sns.relplot(x='CYLINDERS', y='CO2EMISSIONS', hue='FUELTYPE',  data=mydata)
```

```python
# visualizing the relationship between two variables using relplot method
# importing the libraries
import seaborn as sns
import pandas as pd

# loading the dataset
mydata = pd.read_csv('FuelConsumption.csv')

# visualizing the data using relplot
sns.relplot(x='CYLINDERS', y='CO2EMISSIONS', size='CYLINDERS',  data=mydata)
```

```python
# scatter plot using regression line with regplot() method
# importing the libraries
import seaborn as sns

# loading the dataset
mydataset = sns.load_dataset('iris')
mydataset

# visualizing the data using regression line
sns.regplot(x='sepal_length', y='petal_width', data = mydataset)
```

```python
# scatter plot using regression line with lmplot() method
# importing the libraries
import seaborn as sns

# loading the dataset
mydataset = sns.load_dataset('iris')
mydataset

# visualizing the data using regression line
sns.lmplot(x='sepal_length', y='petal_width', data = mydataset)
```

```python
# scatter plot with marginal histograms
# importing the libraries
import seaborn as sns

# loading the dataset
mydata = sns.load_dataset('tips')
mydata

# visualizing the data
sns.jointplot(x='total_bill', y='tip', data = mydata)
```
