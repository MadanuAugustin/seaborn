```python
# seaborn_tutorial
# importing the libraries
import seaborn as sns

# loading the data
mydataset = sns.load_dataset('iris')
mydataset

# visualizing the data
sns.lineplot(x='sepal_length', y='sepal_width', data=mydataset)
```

```python
# using seaborn with pandas
# importing the libraries
import seaborn as sns
import pandas as pd

# loading the data
mydata = pd.read_csv('Iris.csv')
mydata

# visualizing the data
sns.lineplot(x='PetalWidthCm' , y='PetalLengthCm', data = mydata)
```

```python
# using seaborn with matplotlib
# importing the libraries
import seaborn as sns
import matplotlib.pyplot as plt

# loading the data
mydataset_1 = sns.load_dataset('iris')
mydataset_1

# visualizing the data
sns.lineplot(x='sepal_width', y='petal_length', data = mydataset_1)
plt.xlabel('sepal_width')
plt.ylabel('petal_length')
plt.suptitle('Iris_species')
```

```python
# setting the style temporarily using axes_style_method
# importing the libraries
import seaborn as sns

# loading the dataset
mydata = sns.load_dataset('iris')
mydata

# visualizing the data
with sns.axes_style("darkgrid"):
    sns.lineplot(x='sepal_length', y='petal_length', data = mydata)
```
